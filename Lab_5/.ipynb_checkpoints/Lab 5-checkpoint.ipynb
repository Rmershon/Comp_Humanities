## Stemming

```python
cFileDir = "Clean_Books"

uniqWordsD = []
for f in os.listdir(cFileDir):
    docWords = []
    if f.endswith(".txt"):
        fopen = open(cFileDir + "/" + f)

        docWords = set(fopen.read().split())
        uniqWordsD.append(docWords)
```

```python
uniqWordsD
```

Below code will get the stems for each word, but ensure that it is a unique set of them

```python
uniqStemD = []
for i in range(len(uniqWordsD)):
    uniqStem = []
    for word in uniqWordsD[i]:
        s = stemmer.stem(word)
        
        uniqStem.append(s)
    uniqStemD.append(set(uniqStem))
```

```python
docLengthD = {}

for i in range(len(filenames)):
    docLengthD[filenames[i]] = [len(uniqWordsD[i]), len(uniqStemD[i]), (len(uniqWordsD[i]) - len(uniqStemD[i]))]
```

Created a dictionary that contains the filename, then columns of the length of the unique words, stems and their difference. 

```python
import pandas as pd
p = pd.DataFrame(docLengthD)
p = p.T
p.rename(columns={0: 'uniqWords', 1 : 'uniqStems', 2 : 'difference'})
```

We can see that 'The Pickwick Papers' has the greatest difference between unique words and unique stems. The total difference is 5516.

# Lemmatizing

```python
wordListL = ["sleeping", "in", "god", "boats", "walked", "twisting", "apple"]
```

```python
from nltk.corpus import wordnet as wn
```

```python
wn.synsets(wordListL[0])[0]
```

```python
w = wn.synsets(wordListL[0])
```

```python
wordData = {}
for word in wordListL:
    s = wn.synsets(word)[0]
    
    lem = s.lemmas()[0]
    defn = s.definition()
    hyp = s.hyponyms()
    ant = lem.antonyms()
    
    wordData[word] = [lem, defn, hyp, ant]
```

```python
p = pd.DataFrame(wordData).T
p.rename(columns={0: "Lemma", 1:"Definition", 2:"Hyponyms", 3:"Antonyms"})
```

```python
for word in wordData.keys():
    if len(wordData[word][1]) > 0:
        print(word + ": " + wordData[word][1])
```

```python
for word in wordData.keys():
    if len(wordData[word][-1]) > 0:
        print(word, wordData[word][-1])
```

You can see from above, that sleeping and walked have antonyms. All the other words have no antonyms

```python
# For similar words
sim = wn.synsets(wordListL[0])[0].path_similarity(wn.synsets(wordListL[1])[0])
curij = []
for i in range(len(wordListL)):
    wi = wn.synsets(wordListL[i])[0]
    for j in range(i+1, len(wordListL)):
        wj = wn.synsets(wordListL[j])[0]

        newSim = wi.path_similarity(wj)
        if newSim != None:
            if newSim >= sim:
                sim = newSim
                if newSim == sim:
                    curij.append((i, j))
                else:
                    curij = [(i, j)]
                
print(sim, curij)
```

The above code outputs the nearest path found between two words. In this case there are 5 combinations that have the highest value. The combos include

* sleeping and in
* in and god
* god and twisting
* walked and twisting
* walked and apple

```python
highLen = 0
wordL = ''
for word in wordListL:
    w = wn.synsets(word)[0]
    wL = len(list(w.closure(lambda s : s.hypernyms())))
    
    if wL > highLen:
        highLen = wL
        wordL = word
        
print(word, highLen)
```

The word with the greatest number of hyperyms is 'apple' with 15.

# Latent Dirichlet Allocation

```python
from gensim.corpora import Dictionary

#create a Gensim dictionary from the texts
dictionary = Dictionary(sentences)

#remove extremes (similar to the min/max df step used when creating the tf-idf matrix)
dictionary.filter_extremes(no_below=1, no_above=0.8)

#convert the dictionary to a bag of words corpus for reference
corpus = [dictionary.doc2bow(text) for text in sentences]
```

```python
from gensim.models.ldamulticore import LdaMulticore
%time lda = LdaMulticore(corpus, num_topics=5, \
                            id2word=dictionary, \
                            chunksize=10000, \
                            passes=100)
```

```python
topics_matrix = lda.show_topics(formatted=False, num_words=20)
topics_matrix
```

The first topic contains religious words. This makes sense as there are two religious books in the corpus. 

The next topic seems to be heavily related to that of the 'Pickwick Papers'. The is evidence through the words, pickwick, sam, weller, mr, and sir.

Topic 3 is mostly made up of the Eragon books, although the Hobbit appears to be a part of this too. 

The next topic is relating to the two books about King Arthur. With words like arthur and launcelot, sword, damsel, and castle. This topic will probably capture the most middeval books. 

Finally the last topic seems to pretain to Crime and Punishment. Seen through the words like raskolnikov and crime and punishment. 

```python
for f in os.listdir("Clean_Books"):
    if f.endswith(".txt"):
        fopen = open("Clean_Books/" + f)
        bow = dictionary.doc2bow(fopen.read().split())
        print(f, lda.get_document_topics(bow))
```
