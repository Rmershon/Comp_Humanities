```python
def genStats(text):
    uniq = {}
    words = []
    periods = []
    sentences = []
    lett = ''
    chars = ''
    curSent = 0
    #textLines = text.readlines()
    
    for line in text:
        cur = line.replace('-', ' ')
        cur = re.sub(r'[^a-zA-Z\s]', '', cur).replace('\n', '').lower()
        lett += cur.replace(' ', '')
        words += cur.split(' ')
        chars += re.sub(r"[^\S]", '', line).replace('\n', ' ').lower()
        periods += re.sub(r'[^.?!]', '', line)
       
    for word in words:
        if word not in ['\x0cit', '', '\x0cthat', '\x0c']:
            if word in uniq.keys():
                uniq[word] += 1
            else:
                uniq[word] = 1
        else: 
            del word
    
    sent = len(periods)
    
    
    
    
    return chars, lett, sent, words, uniq
```

```python
class Word:
    
    def __init__(self, x, y, text, size, rot):
        self.initx = x
        self.inity = y
        self.r = 0
        self.theta = 0
        
        self.x = x
        self.y = y
        self.text = text
        self.size = size
        
        self.rot = rot
        
    def right(self):
        if self.rot: 
            return self.x + self.size * 1/72
        return self.x + len(self.text) * .75 * self.size * 1/72
        
    def top(self):
        if self.rot:
            return self.y + len(self.text) * .75 * self.size * 1/72
        return self.y + self.size * 1/72
        
#     def right(self):
#         return self.x + len(self.text) * .75 * self.size * 1/72
        
#     def top(self):
#         return self.y + self.size * 1/72
        
    def overlap(self, other):
        # http://gamemath.com/2011/09/detecting-whether-two-boxes-overlap/
        if self.right() < other.x:
            return False
        if other.right() < self.x:
            return False
        if self.top() < other.y:
            return False
        if other.top() < self.y:
            return False
        return True
    
    def move(self):
        self.r += 0.01
        self.theta += 0.3
        newx, newy = pol2cart(self.r, self.theta)
        self.x = self.initx + newx
        self.y = self.inity + newy
```

```python
#https://stackoverflow.com/questions/20924085/python-conversion-between-coordinates

def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return(rho, phi)

def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)
```
